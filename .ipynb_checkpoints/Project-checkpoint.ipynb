```python
#Libraries needed for this project
import openpyxl
import time
import numpy as np
import requests
from datetime import datetime, date
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
import urllib
from urllib import error
```

```python
def corona_update():
    try:
        #while loop to self-automate the program, once your batch program is ran
        while True:
            #To get access to the ncdc website 
            urlopen('http://216.58.192.142', timeout=1)
            # print(f'Covid 19 Nigerian daily numbers update')
            response = requests.get("https://covid19.ncdc.gov.ng/")
            soup = BeautifulSoup(response.text, "html.parser")

            #To get the data from the h2 class div "text-right text-white", where the datra we want is stored
            cases = soup.find_all("h2", class_="text-right text-white")
            # print(soup.title.text)
            confirmed_cases = cases[1].getText()
            active_cases = cases[2].getText()
            deaths = cases[4].getText()

            #funtions to store all three data we need
            def confirmedcases():
                return confirmed_cases

            def activecases():
                return active_cases

            def death():
                return deaths

            
            #dt variable is used to set the date format
            dt = datetime(2020, 12, 23)
            #dt_now variable stores the current date
            dt_now = datetime.now().date()
            #The translate function replaces "," with none 
            case1 = confirmedcases().translate({ord(','): None})
            # case12 = case1.translate({ord(','): None})
            case2 = activecases().translate({ord(','): None})
            case3 = death().translate({ord(','): None})


            #The inputs varibale stores all the values in a numpy array
            inputs = np.array([dt_now, case1, case2, case3])


            #This is to open the excel workbook sheet
            wb = openpyxl.load_workbook("CORONA_VIRUS_DATA.xlsx")
            print(wb.sheetnames)
            sheet = wb["Sheet1"]

            #row is the text file, which countains the row number where the new values are to be stored.
            with open("row.txt", "r") as file:
                file.seek(0)
                num = file.readline()
                print(num)

            #This is to get the row number where the last data was stored.
            row = int(num)
            col = 1
            row = row - 1

            #This is to get the date of the row the last data was stored.
            cell = sheet.cell(row=row, column=1)
            excel_date = cell.value
            excel_date2 = datetime.date(excel_date)

            # py_excel_date = datetime.fromordinal(datetime(1900, 1, 1).toordinal() + excel_date - 2)
            # py_excel_date2 = datetime.date(py_excel_date)

            print(excel_date2)
            print(dt_now)
            #This is to verify that the current date "dt_now" is greater than the last date where data was stored "excel_date2"
            if excel_date2 < dt_now:
                #This is to get the row number, where the values for today is to be stored.
                row = row + 1

                #This is to insert the four variables in the numpy array into the columns, respectively
                for number in range(4):
                    cell = sheet.cell(row=row, column=col)
                    cell.value = inputs[number]
                    col += 1
                    print(cell.value)
                    

                #This is to save the excel workbook changes
                wb.save("CORONA_VIRUS_DATA.xlsx")

                #This is to change the excel file format to csv
                data_xlsx = pd.read_excel('CORONA_VIRUS_DATA.xlsx', 'Sheet1', index_col=None)

                #This is to save the new csv format into a csv file
                data_xlsx.to_csv('CORONA_VIRUS_DATA.csv', encoding='utf-8')

                # wb.save("CORONA_VIRUS_DATA.csv")
                print("Your data has been updated.")

                #This is to change the row variables value to the row number, where the values for the next day is to be stored.
                srow = str(row + 1)

                with open("row.txt", "w") as file:
                    file.write(srow)
            elif excel_date2 > dt_now:
                print("Date is invalid, please confirm.")
            elif excel_date2 == dt_now:
                print("You're updated.")

            #This complements the while loop, and lets the program sleep for 1 hour before its ran again
            time.sleep(3600)

    #This is an exception handler that tells the program, if ran into tells the program sleep for 300 seconds before proceeding.
    except urllib.error.URLError as ex:
        print(ex)
        time.sleep(300)
        corona_update()
        
    #This is an exceptionhandler that tells the user what to do, if ran into.
    except PermissionError as ex:
        print(type(ex))
        print(ex)
        print("Close your excel, text or csv document.")

#This calls the function that has been defined above.
corona_update()
```
